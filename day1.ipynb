```python
def number_triangle(n):
    for i in range(1, n + 1):
        for j in range(1, i + 1):
           print(j, end=' ')
        print()

# Example Usage:
number_triangle(5)
```

```python
def reverse_number_triangle(n):
    for i in range(n, 0, -1):
        for j in range(1, i + 1):
           print(j, end=' ')
        print()

# Example Usage:
reverse_number_triangle(5)
```
