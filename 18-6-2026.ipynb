```python
# function defination

def percent (phy, maths, maxmarks = 200):
    val = (phy + maths) * 100/maxmarks
    return val

phy = 60
maths = 70

# function calling with default argument

result = percent(phy, maths)
print("percentage:", result)
```

```python
def fcn (nums, numericlist = []):
    numericlist.append(nums + 10)
    print(numericlist)

# function calls

fcn(67)
fcn(657)
```

```python
def right_angled_triangle(n):
    for i in range(1, n + 1):
        print("+ " * i)

# example usage:
right_angled_triangle(11)
```

```python
def inverted_right_angled_triangle(n):
    for i in range(n, 0, - 1):
        print("+ " * i)

# example usage:
inverted_right_angled_triangle(11)
```

```python
def pyramid_pattern(n):
    for i in range(1, n + 1):
        print(' ' *(n - i) + '*' * (2 * i - 1))

# example usage:
pyramid_pattern(11)
```

```python
def pyramid_pattern(n):
    for i in range(1, n + 1):
        print(' ' *(n - i) + '|' * (2 * i - 1))
    for i in range(n - 1, 0, - 1):
        print(' ' *(n - i) + '|' * (2 * i - 1))

# example usage:
pyramid_pattern(11)
```
